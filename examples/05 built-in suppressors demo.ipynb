```python
from pyIClab._testing_toolkit import PackedIC
from pyIClab import __version__ as pyIClab__version__
from pyIClab import (
    QuickSuppressor, PhreeqcSuppressor, InsufficientPhreeqcSuppressor,
    ContaminatedPhreeqcSuppressor, Eluent)

pyIClab__version__
```

```python
supressor_args = {'name': 'Supressor', 'kind': 'anion'}
suppressors = [
    QuickSuppressor(**supressor_args),
    PhreeqcSuppressor(**supressor_args),
    InsufficientPhreeqcSuppressor(**supressor_args),
    ContaminatedPhreeqcSuppressor(**supressor_args),
    ]
solution = {
    'F-': '0.005mM',
    'Cl-': '0.0075mM',
    'Br-': '0.0125mM',
    'NO3-': '0.0125mM',
    }

ic1, ic2, ic3, ic4 = [
    PackedIC(
        competing_ions=('OH-',), 
        eluent=Eluent(name='EG', profile={'OH-': [('4 min', '10 mM'), ('10 min', '55 mM')]}),
        suppressor=suppressors[i],
        sample_solution=solution
        ) for i in range(4)
    ]


for ic in (ic1, ic2, ic3, ic4):
    
    ic.go('10 min')
```

```python
signals = []
for ic in [ic1, ic2, ic3, ic4]:
    (detector,) = tuple(ic.detectors)
    df = detector.get_signals(signal_type='conductivity')
    signals.append(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from pyIClab.beadedbag import mpl_custom_rcconfig

sns.set()
plt.rcParams.update(mpl_custom_rcconfig)
fig, ax = plt.subplots()

for i, ic in enumerate([ic1, ic2, ic3, ic4]):
    df = signals[i]
    (suppressor,) = tuple(ic.suppressors)
    label = type(suppressor).__name__
    ax.plot(df['time'], df['signal'] + 0.05*i, label=label)
ax.legend()
ax.set(xlim=(0, 10))
ax.set_xlabel('t, min', fontsize=10, fontweight='bold')
ax.set_ylabel('conductivity, μS/cm', fontsize=10, fontweight='bold')
```
