```python
from pyIClab import Eluent
from pyIClab import __version__ as pyIClab__version__

pyIClab__version__
```

```python
eluent = Eluent.HydroxideIsocratic('15 mM')
eluent.plot()
```

```python
eluent = Eluent.HydroniumIsocratic('15 mM')
eluent.plot()
```

```python
eluent = Eluent.Carbonates(carbonate=2.5, bicarbonate=1.2)
eluent.plot()
```

```python
eluent = Eluent('KOHGradient',
    profile={'hydroxide': (
        ('5 min', '5 mM'),
        ('10 min', '24 mM'),
        ('15 min', '24 mM'),
        ('30 min', '55 mM'),
        ('30 min', '5 mM'),
        )}, 
    fr='1.25 mL/min')
eluent.plot()
```

```python
eluent = Eluent('WontBeUseful',
    profile={
        'acetate': ((5, 5), (7, 12)),
        'Mg+2': '11 mM',
        'NoFormula[-2]': ((12, 2.5), (17, 5)),
        'SO4-2': '0.003 mol/L',})
eluent.plot()
```

```python
import numpy as np
eluent = Eluent.from_funcs('YouNameIt',
    profile={
        'OH-': lambda t: np.e**(0.25*t) + 0.25,
        'SO4-2': lambda t: 2*np.sin(t)**2 + 1,
        },
    tmax='10 min',
    )
eluent.plot()
```
